```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x+1

x=np.arange(-5, 6)
y=f(x)
plt.plot(x, y)
```

f(x) = x*2 + 2x+1
x의 법위는 -5에서 5까지

```python
def f1(x):
    return 2*x+1

def f2(x):
    return x**2 + 2*x +1
# x = np.arange(-5,5)
x = np.linspace(-5,5,100) # 간격을 넣어서, 좀 더 부드러운 곡선이 됨
y1 = f1(x)
y2 = f2(x)

plt.plot(x,y1,c='yellow', ls='--', lw=5, label='f1(x)') # c='' 색상변경/ ls='' 라인스타일/ lw='' 굵기/ label='' 
plt.plot(x,y2,c='red', lw=5, label='f2(x)') 
plt.grid() # 격자
plt.legend() # label넣으면 범례가 생김
plt.title('Two Graphs') #제목
plt.xlabel('X') # x축 이름
plt.ylabel('Y') # y축 이름
plt.savefig('p.png')
```

```python
from random import choice
class RandomWalk:
    def __init__(self, num_points=5000):
        self.num_points = num_points
        self.x_values = [0]
        self.y_values = [0]
    def fill_walk(self):
        while len(self.x_values) < self.num_points:
            x_direction = choice([1,-1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction * x_distance

            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction * y_distance
            if x_step == 0 and  y_step == 0:
                continue
            x=self.x_values[-1] + x_step
            y=self.y_values[-1] + y_step

            self.x_values.append(x)
            self.y_values.append(y)
```

```python
rw = RandomWalk()
rw.fill_walk()

plt.style.use('classic')
plt.scatter(rw.x_values, rw.y_values,s=1, c=range(5000), cmap=plt.cm.viridis, edgecolor='none')
```

```python
from random import randint

class Die:
    def __init__(self, num_sides=6):
        self.num_sides = num_sides
    def roll(self, time=1):
        return [randint(1, self.num_sides) for _ in range(100)]
   
```

```python
d1 =Die()
result = d1.roll(100)
print(result)
plt.hist(result, bins=6, edgecolor='black')
```
