실습- 특성간 관계 분석

```python
iris = load_iris()
X = iris.data #입력데이터 독립변수
y = iris.target #품종 종속변수 
feature_names = iris.feature_names
target_names = iris.target_names
```

```python
# 한국어폰트(윈도우)
plt.rcParams["font.family"] = "Malgun Gothic"
# 맥 => AppleGothic

plt.rcParams["axes.unicode_minus"] = False
```

```python
df = pd.DataFrame(X, columns = feature_names)
df["species"] = [str(iris.target_names[i]) for i in y]
```

```python
feature_names_kr=["꽃받침 길이","꽃받침 너비","꽃잎 길이","꽃잎 너비"]
target_names_kr=["세토사","버시컬러","버지니카"]
```

```python
# 시각자료 만들때 컬러먼저 세팅하기
colors = ["#FF6B6B", "#4ECDC4","#45B7D1"]
species_colors = {target_names[i] : colors[i] for i in range(3)}
```

특성 간 관계분석

```python
# 분석적 목적: 특성 간 분포와 분리 가능성 파악
# 16×16 크기의 그림 공간을 만들고, 그 안에 **4행 4열(for문 i,j)의 서브플롯(작은 그래프)**을 생성
fig, axes = plt.subplots(4,4, figsize=(16,16))
fig.suptitle("Iris 데이터 특성 간 관계 분석", fontsize=50);

# (4,4)
# 16개중 첫번째 그래프 = (0,0)
print(axes.shape)

# (0,0)~(3,3)
for i in range(4):
    for j in range(4):
        print(i,j)

# 지정 그래프 가져오기 : (i,j)로 (0,0)가져오기
for i in range(4):
    for j in range(4):
        ax = axes[i, j]

        if i == j:
            # 막대그래프! => 도수 분포를 위한 '데이터'를 정리해야함
            # enumerate : 값, 인덱스
            for k, species in enumerate(target_names):
                data = df[df["species"] == species].iloc[:, i]
                ax.hist(
                    data, 
                    color=colors[k], 
                    bins=12, 
                    density=True,  # 전체 면적(막대 면적 합)이 1이 되도록
                    alpha=0.6, 
                    label=target_names_kr[k]
                )
            ax.set_title(feature_names_kr[i])
            ax.set_ylabel("밀도")
            if i == 0:
                ax.legend()

        else:
            for k, species in enumerate(target_names):
                # 산점도
                data = df[df["species"] == species]

                # 주대각선 기준으로 대칭이 되어야 함 (i, j 선택)
                x = data.iloc[:, j]  # 독립변인 (열)
                y = data.iloc[:, i]  # 종속변인 (행)

                ax.scatter(
                    x,
                    y,
                    color=colors[k],
                    alpha=0.6,
                    label=target_names_kr[k],
                    s=30,  # 점 사이즈
                    edgecolors="white"  # 점 윤곽선
                )

            ax.set_xlabel(feature_names_kr[j])
            ax.set_ylabel(feature_names_kr[i])

            corr = df.iloc[:, [j,i]].corr().iloc[0,1]
            # 그래프 안 텍스트위치 p.309
            ax.text(
                0.05,
                0.95,
                f"r={corr:.2f}",
                transform = ax.transAxes
            )
            ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
